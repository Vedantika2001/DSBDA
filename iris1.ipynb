```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
x = df['Species']
y = df['SepalWidthCm']
plt.ylabel("SepalWidthCm")
plt.title('graph')
plt.bar(x,y,label='width')
plt.legend()
```

```python
df['SepalLengthCm'].hist(bins=70)
```

```python
a=['1300','450','760']
b=['virginica','sefosa','versicolor']
plt.pie(a,labels=b, radius=2, autopct='%0.0f%%')
```

```python
df.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
```

```python
sns.boxplot(x='Species', y = 'SepalLengthCm', data=df)
```

```python
sns.pairplot(df)
plt.title('Pairplot for the Data', fontsize = 20)
plt.show()
```
